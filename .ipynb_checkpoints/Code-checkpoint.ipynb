  <h3 align="center">2. Data Preprocessing and Cleaning</h3> 

```python
(trf.isnull().sum())  #Null values check
```

```python
trf_cleaned= trf.dropna(subset=['STREET_NAM']) #Based on the null values droppped values in Street names
```

```python
trf_cleaned.isnull().sum()  #Checked once again(Null Values)
```

```python
'''trf_cleaned['AB_FLOW_AM'] = trf_cleaned['AB_FLOW_AM'].astype(float)'''  # To change the datatype just in case 
trf_cleaned.dtypes 

trf_cleaned['AB_CONGESTION_AM'] = trf_cleaned['AB_FLOW_AM'] / trf_cleaned['AB_CAPACI1'] # Calculate congestion ratio (flow/capacity)

trf_cleaned['TRAVEL_TIME_AB'] = trf_cleaned['LENGTH'] / trf_cleaned['FF_SPEED'] * 60  # Calculate travel time in minutes (Length in km / Speed in km/h)


trf_cleaned['AVG_DAILY_FLOW'] = (trf_cleaned['AB_VOL_DAI'] + trf_cleaned['BA_VOL_DAI']) / 2 # Calculate average daily flow


trf_cleaned['TRUCK_RATIO'] = trf_cleaned['TOT_TRK_DA'] / trf_cleaned['TOT_VOL_DA'] # Calculate Truck to Total Volume Ratio


trf_cleaned['TVCR_AM'] = trf_cleaned['AB_FLOW_AM'] / trf_cleaned['AB_CAPACI1'] # Calculate Volume to Capacity Ratio for AM Peak Hours


trf_cleaned['TVCR_PM'] = trf_cleaned['AB_FLOW_PM'] / trf_cleaned['AB_CAPACI1'] # For PM Peak Hours

```

```python
                    # Categorize congestion levels
import numpy as np
conditions = [
    (trf_cleaned['AB_CONGESTION_AM'] <= 0.5),  # Low congestion
    (trf_cleaned['AB_CONGESTION_AM'] > 0.5) & (trf_cleaned['AB_CONGESTION_AM'] <= 1),  # Moderate congestion
    (trf_cleaned['AB_CONGESTION_AM'] > 1)  # High congestion
]
categories = ['Low', 'Moderate', 'High']

trf_cleaned['Congestion_Category'] = np.select(conditions, categories)
```

#### Route Optimization and Traffic Management

For roads with high congestion, analyze alternative routes that have lower traffic volumes and higher capacities.

You can use shortest path algorithms (like Dijkstra's algorithm) to recommend optimized routes.

```python
import networkx as nx

# Create a graph of the transport network using NetworkX
G = nx.Graph()

# Use the correct column for street names
for index, row in trf_cleaned.iterrows():
    G.add_edge(row['STREET_NAM'], row['STREET_NAM'], length=row['LENGTH'], capacity=row['AB_CAPACI1'] + row['BA_CAPACI1'])

# Verify if the streets exist in the graph
print('BUCKMINSTER ROAD' in G)  # Should return True if 'BUCKMINSTER ROAD' exists in the graph
print('SUMNER ROAD' in G)  # Should return True if 'SUMNER ROAD' exists in the graph

# Check if the graph is connected
is_connected = nx.is_connected(G)
print(f"Is the graph connected? {is_connected}")

# If the graph is not connected, find the number of components
if not is_connected:
    components = list(nx.connected_components(G))
    print(f'Number of connected components: {len(components)}')

    # Optionally, print the nodes in each disconnected component
    for i, component in enumerate(components):
        print(f"Component {i}: {component}")

# Check the neighbors of 'BUCKMINSTER ROAD'
print(f"Neighbors of BUCKMINSTER ROAD: {list(G.neighbors('BUCKMINSTER ROAD'))}")

# Check the neighbors of 'SUMNER ROAD'
print(f"Neighbors of SUMNER ROAD: {list(G.neighbors('SUMNER ROAD'))}")

# Try calculating the shortest path
try:
    shortest_path = nx.shortest_path(G, source='BUCKMINSTER ROAD', target='SUMNER ROAD', weight='length')
    print(f'Shortest Path: {shortest_path}')
except nx.NetworkXNoPath as e:
    print(f'No path between BUCKMINSTER ROAD and SUMNER ROAD: {e}')
except nx.NodeNotFound as e:
    print(f'Node not found: {e}')
```

Graph Creation: We're building a map (graph) where each street is a point (node) and connections between streets are lines (edges) with attributes like road length.


Adding Connections: Each street is added to the map, and the roads between them are given distances (lengths).


Checking Streets: The program checks if specific streets (e.g., BUCKMINSTER ROAD and SUMNER ROAD) are part of the map.


Shortest Path Calculation: The program finds the shortest route between two streets by adding up the distances between streets and picking the shortest option.


Handling Errors: If there is no connection between the streets, the program lets us know by raising an error.


Output: If a path exists, it prints the shortest route; otherwise, it informs you that no path exists.

### Machine Learning Models

We applied multiple machine learning models to compare their performance and identify which one best predicts traffic volume based on road data. Different models capture patterns in data differently, so comparing them ensures we find the most accurate and reliable model. This helps improve the accuracy of traffic predictions and provides better insights for decision-making.








```python
import matplotlib.pyplot as plt
from sklearn.ensemble import RandomForestRegressor, GradientBoostingRegressor
from sklearn.svm import SVR
from sklearn.neural_network import MLPRegressor
from sklearn.metrics import mean_squared_error, r2_score
import xgboost as xgb

# List to store the results of each model
results = []

# 1. Random Forest
rf_model = RandomForestRegressor(n_estimators=100, random_state=42)
rf_model.fit(X_train, y_train)
rf_predictions = rf_model.predict(X_test)
rf_mse = mean_squared_error(y_test, rf_predictions)
rf_r2 = r2_score(y_test, rf_predictions)
results.append(('Random Forest', rf_mse, rf_r2))

# 2. Gradient Boosting Regressor
gbr_model = GradientBoostingRegressor(n_estimators=100, random_state=42)
gbr_model.fit(X_train, y_train)
gbr_predictions = gbr_model.predict(X_test)
gbr_mse = mean_squared_error(y_test, gbr_predictions)
gbr_r2 = r2_score(y_test, gbr_predictions)
results.append(('Gradient Boosting', gbr_mse, gbr_r2))

# 3. XGBoost
xg_model = xgb.XGBRegressor(n_estimators=100, random_state=42)
xg_model.fit(X_train, y_train)
xg_predictions = xg_model.predict(X_test)
xg_mse = mean_squared_error(y_test, xg_predictions)
xg_r2 = r2_score(y_test, xg_predictions)
results.append(('XGBoost', xg_mse, xg_r2))

# 4. Support Vector Regressor (SVR)
svr_model = SVR(kernel='rbf')
svr_model.fit(X_train, y_train)
svr_predictions = svr_model.predict(X_test)
svr_mse = mean_squared_error(y_test, svr_predictions)
svr_r2 = r2_score(y_test, svr_predictions)
results.append(('Support Vector Regressor', svr_mse, svr_r2))

# 5. MLP Regressor (Neural Network)
mlp_model = MLPRegressor(hidden_layer_sizes=(100, 50), max_iter=500, random_state=42)
mlp_model.fit(X_train, y_train)
mlp_predictions = mlp_model.predict(X_test)
mlp_mse = mean_squared_error(y_test, mlp_predictions)
mlp_r2 = r2_score(y_test, mlp_predictions)
results.append(('MLP Regressor', mlp_mse, mlp_r2))

# Convert results to lists for plotting
models = [x[0] for x in results]
mse_scores = [x[1] for x in results]
r2_scores = [x[2] for x in results]

# Plotting Mean Squared Error (MSE)
plt.figure(figsize=(10, 5))

plt.subplot(1, 2, 1)
plt.barh(models, mse_scores, color='skyblue')
plt.title('Mean Squared Error (MSE) Comparison')
plt.xlabel('MSE')
plt.ylabel('Models')

# Plotting R-squared
plt.subplot(1, 2, 2)
plt.barh(models, r2_scores, color='lightgreen')
plt.title('R-squared (R2) Comparison')
plt.xlabel('R-squared')
plt.ylabel('Models')

plt.tight_layout()
plt.show()
```

Random Forest and Gradient Boosting show the best performance with lower MSE and better (though still negative) R-squared values.
XGBoost performs reasonably but slightly worse than Random Forest and Gradient Boosting.
MLP Regressor and SVR perform poorly with high MSE and highly negative R-squared values.
